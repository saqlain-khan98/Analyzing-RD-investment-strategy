# Analysis of Investing in an RD in INR with 7% interest from EUR.
 To simulate a more lifelike behavior of exchange rates, we could use techniques such as Monte Carlo simulations, where we introduce randomness to mimic real-world fluctuations. These fluctuations could include gradual trends, small random variations, and abrupt changes.

##### Steps:
- Historical Data Integration: We'll base the simulations on the statistical behavior of historical EUR to INR exchange rates, like volatility and drift.
- Monte Carlo Simulation: Simulate 1000 possible future exchange rate paths over the next 10 years, factoring in random noise, drift, and potential shocks.
-Data Source: For accurate modeling, I can integrate real-world data on EUR to INR from the past several years to estimate parameters like volatility and drift.

##### Components of the Simulation:
- Drift: Long-term trend based on historical data.
- Volatility: Day-to-day fluctuations.
- Shocks: Occasional, abrupt changes.

```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Load the historical data provided by the user
file_path = "/Users/khan/Downloads/EUR_INR Historical Data.xls"
historical_data = pd.read_csv(file_path)

# Display the first few rows to understand the structure
historical_data.head()
# Convert the 'Date' column to datetime
historical_data['Date'] = pd.to_datetime(historical_data['Date'], format='%m/%d/%Y')

# Sort the data by date in ascending order
historical_data = historical_data.sort_values('Date')

# Calculate daily returns (percentage changes in price)
historical_data['Price'] = historical_data['Price'].astype(float)
historical_data['Daily_Return'] = historical_data['Price'].pct_change()

# Remove the first row which will have NaN in 'Daily_Return'
historical_data = historical_data.dropna(subset=['Daily_Return'])

# Calculate statistical parameters: mean daily return (drift) and volatility
drift = historical_data['Daily_Return'].mean()
volatility = historical_data['Daily_Return'].std()

# Display calculated drift and volatility
drift, volatility
# Load interest rate data
ecb_rates_path = "/Users/khan/Downloads/ECB_rate.csv"
rbi_rates_path = "/Users/khan/Downloads/RBI_rate.csv"
ecb_rates = pd.read_csv(ecb_rates_path)
rbi_rates = pd.read_csv(rbi_rates_path)

# Convert 'Date' column to datetime in interest rate data
ecb_rates['Date'] = pd.to_datetime(ecb_rates['Date'], format='%m/%d/%Y')
rbi_rates['Date'] = pd.to_datetime(rbi_rates['Date'], format='%d-%m-%Y')

# Ensure unique dates in interest rate data
ecb_rates = ecb_rates.drop_duplicates(subset='Date')
rbi_rates = rbi_rates.drop_duplicates(subset='Date')
rbi_rates['RBI_Repo_Rate'] = rbi_rates['RBI_Repo_Rate'].str.rstrip('%').astype('float')
# Set 'Date' as the index for forward filling
ecb_rates.set_index('Date', inplace=True)
rbi_rates.set_index('Date', inplace=True)

# Create a date range that spans the entire period of historical data
date_range = pd.date_range(start=historical_data['Date'].min(), end=historical_data['Date'].max(), freq='D')

# Reindex the interest rate data to match the full date range and forward fill
ecb_rates = ecb_rates.reindex(date_range).ffill().reset_index().rename(columns={'index': 'Date'})
rbi_rates = rbi_rates.reindex(date_range).ffill().reset_index().rename(columns={'index': 'Date'})

# Plot the historical data
plt.figure(figsize=(12, 6))
plt.plot(historical_data['Date'], historical_data['Price'], color='green', label='EUR/INR Exchange Rate')

# Add titles and labels
plt.title('Historical EUR/INR Exchange Rate', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Exchange Rate (INR per 1 EUR)', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add grid and legend
plt.grid(True)
plt.legend()

# Show the plot
plt.show()


# Create a figure with 2 subplots (1 row, 2 columns)
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot the ECB data in the first subplot
axes[0].plot(ecb_rates['Date'], ecb_rates['ECB_Rate'], color='blue', label='ECB Interest Rate')
axes[0].set_title('ECB Interest Rate', fontsize=16)
axes[0].set_xlabel('Date', fontsize=12)
axes[0].set_ylabel('Interest Rate', fontsize=12)
axes[0].grid(True)
axes[0].legend()
axes[0].tick_params(axis='x', rotation=45)  # Rotate x-axis labels

# Plot the RBI data in the second subplot
axes[1].plot(rbi_rates['Date'], rbi_rates['RBI_Repo_Rate'], color='goldenrod', label='RBI Interest Rate')
axes[1].set_title('RBI Interest Rate', fontsize=16)
axes[1].set_xlabel('Date', fontsize=12)
axes[1].set_ylabel('Interest Rate', fontsize=12)
axes[1].grid(True)
axes[1].legend()
axes[1].tick_params(axis='x', rotation=45)  # Rotate x-axis labels

# Adjust layout to avoid overlapping
plt.tight_layout()

# Show the plot
plt.show()
```

## Simulation of interest rates.

Interest rates of ECB are simulated by keeping in mind the effect of corona pandemic. Due to the pandemic the ECB had risen interest rate which let to slowing of economy but decrease in inflation rates. As the ECB has driven inflation rates to its desired 2% mark, it plans to lower interest rates. It is assumed that ECB will change its interest rates slowly.

The RBI aims to keep the INR value stable. Hence RBI would also change interest rates by keeping a margin of 4.0% higher interest rates compared to ECB. 

```python
# Get the last known rates
ecb_start_rate = ecb_rates['ECB_Rate'].iloc[-1]
rbi_start_rate = rbi_rates['RBI_Repo_Rate'].iloc[-1]

# Simulation parameters
target_ecb_rate = 0.5  # Target rate to maintain 2% inflation
num_days = 365.25 * 10 # 10 years
ecb_dates = pd.date_range(start=pd.to_datetime(ecb_rates['Date'].iloc[-1]) + pd.DateOffset(days=1),
                          periods=num_days + 1, freq='D')
rbi_dates = ecb_dates  # Same dates since they are aligned

# Simulate ECB interest rates
ecb_simulated_rates = []
rbi_simulated_rates = []
num_month = ecb_dates[0].month
for i in ecb_dates:
    if i == ecb_dates[0]:
        ecb_simulated_rates.append(ecb_start_rate)
        rbi_simulated_rates.append(rbi_start_rate)        
    elif i.month == num_month:
        ecb_simulated_rates.append(ecb_simulated_rates[-1])
        rbi_simulated_rates.append(rbi_simulated_rates[-1])
    elif (i.month % 6) == 0: 
        # Randomly decide to increase ECB rate due to shocks
        if np.random.rand() < 0.1:  # 10% chance of a shock
            ecb_shock_increase = 0.25  # Increase by 25% during shock
            new_ecb_rate = ecb_simulated_rates[-1] + ecb_shock_increase
        else:
            # Update ECB rate based on previous logic
            if ecb_simulated_rates[-1] > target_ecb_rate:
                new_ecb_rate = max(ecb_simulated_rates[-1] - 0.125, target_ecb_rate)  # Decrease by 0.015% per month
            else:
                new_ecb_rate = ecb_simulated_rates[-1]  # Stabilize at target

        ecb_simulated_rates.append(new_ecb_rate)

        # Update RBI rate to be 3.5% higher than ECB rate
        new_rbi_rate = new_ecb_rate + 4.0
        rbi_simulated_rates.append(new_rbi_rate)
        if num_month < 12:
            num_month + 1
        else:
            num_month = 0
    else:
        ecb_simulated_rates.append(ecb_simulated_rates[-1])
        rbi_simulated_rates.append(rbi_simulated_rates[-1])
        if num_month < 12:
            num_month + 1
        else:
            num_month = 0

            
# Create DataFrames for plotting
ecb_df = pd.DataFrame({'Date': ecb_dates, 'ECB_Rate': ecb_simulated_rates})
rbi_df = pd.DataFrame({'Date': rbi_dates, 'RBI_Repo_Rate': rbi_simulated_rates})

# Plotting ECB interest rates
plt.figure(figsize=(12, 6))
# First plot for ECB interest rate (subplot 1 of 2)
plt.subplot(1, 2, 1)  # (1 row, 2 columns, 1st subplot)
plt.plot(ecb_df['Date'], ecb_df['ECB_Rate'], color='blue', label='Simulated ECB Interest Rate')
plt.axhline(y=target_ecb_rate, color='gray', linestyle='--', label='Target ECB Rate (2% Inflation)')
plt.title('Simulated ECB Interest Rate for Next 10 Years', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Interest Rate (%)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
# Second plot for RBI interest rate (subplot 2 of 2)
plt.subplot(1, 2, 2) 
plt.plot(rbi_df['Date'], rbi_df['RBI_Repo_Rate'], color='orange', label='Simulated RBI Repo Rate')
plt.axhline(y=rbi_start_rate + 3.5, color='gray', linestyle='--', label='Target RBI Rate (4.0% above ECB)')
plt.title('Simulated RBI Repo Rate for Next 10 Years', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Interest Rate (%)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
#print(len(ecb_dates))
```

## Monte Carlo simulations of the EUR to INR exchange rates.
Performing a range of simulations would help in understanding how would different strategies turn up for different scenarios. This will help in to analyze the strategies and their dependence in the exchange rates.

Monte Carlo simulations are perfomed to mimic different scenarios of the exchange rates for the duration of next ten years. The simulation take volatility and drift of the exchange rates from the available historical data. The simulations also generate a random shock based on drift and volatility. To have a wide range of scenarios, ten thousand simulations are performed. Twenty five randomly selected scenarios are plotted in the following graph. 

```python
# Parameters for the simulation
num_simulations = 1000
num_months = 120
initial_rate = historical_data['Price'].iloc[-1]  # Starting from the most recent price
time_steps = np.arange(1, num_months + 1)


# Parameters for the recurring deposit (RD)
monthly_inr = 60000  # INR
interest_rate = 0.07  # 7% annual interest
monthly_interest_rate = interest_rate / 12  # Monthly interest rate

# Simulate exchange rates using Monte Carlo
np.random.seed(42)  # For reproducibility

# Preallocate array for all simulations
simulated_rates = np.zeros((num_months, num_simulations))

# Simulate each scenario
for i in range(num_simulations):
    simulated_rates[0, i] = initial_rate
    for t in range(1, num_months):
        # Generate a random shock based on drift and volatility
        random_shock = np.random.normal(drift, volatility)
        simulated_rates[t, i] = simulated_rates[t-1, i] * (1 + random_shock)

# Calculate the total EUR spent for each scenario
eur_spent_simulations = np.array([np.sum( monthly_inr / simulated_rates[:, i]) for i in range(num_simulations)])

# Plotting the first few simulations for clarity
plt.figure(figsize=(12, 8))

# Plot a subset of simulations (e.g., 10 out of 1000)
for i in random.sample(range(num_simulations), 25):
    plt.plot(time_steps, simulated_rates[:, i], lw=0.7) # , label=f'Simulation {i+1}' if i <  else "")
    plt.text(num_months, simulated_rates[-1, i], f'{eur_spent_simulations[i]:.2f} EUR', fontsize=8, color='red')

# Add labels and title
plt.xlabel('Months (Time)')
plt.ylabel('Exchange Rate (INR per 1 EUR)')
plt.title('25 Monte Carlo Simulations of EUR to INR Exchange Rate Over 10 Years')

# Mark total EUR spent at the end of each selected simulation
#for i in range(10):
#    plt.text(num_months, simulated_rates[-1, i], f'{eur_spent_simulations[i]:.2f} EUR', fontsize=8, color='red')

# Show the plot
plt.grid(True)
#plt.legend(loc='upper left')
plt.show()
```

## RD investment strategy

In 2024, HDFC bank offers a recurring deposit of with interest rate of 7.0%. The following graph shows how a monthly RD of Rs.60,000 will lead to a lumpsum of nearly 1 crore rupees. 

```python

# Function to calculate the growth of the recurring deposit and interest gained
def calculate_rd_growth_and_interest(monthly_inr, num_months, monthly_interest_rate):
    total_amount = 0
    deposit_growth = []
    interest_gained = []
    invested_amount = []
    for month in range(1, num_months + 1):
        previous_amount = total_amount
        total_amount += monthly_inr
        total_amount *= (1 + monthly_interest_rate)
        deposit_growth.append(total_amount)
        interest_gained.append(total_amount - (previous_amount + monthly_inr))  # Interest gained in this month
        invested_amount.append(monthly_inr * month)  # Total principal invested until this month
    return deposit_growth, interest_gained, invested_amount

# Calculate the growth of the recurring deposit and interest gained
rd_growth, interest_gained, invested_amount = calculate_rd_growth_and_interest(monthly_inr,
                                                                               num_months, monthly_interest_rate)

# Create a stacked bar chart
plt.figure(figsize=(12, 6))

# x-axis values: months (1 to 120)
x = np.arange(1, num_months + 1)

# Stacked bar chart: invested amount + interest gained
plt.bar(x, invested_amount, label='Invested Amount (Principal)', color='darkgoldenrod')
plt.bar(x, np.array(rd_growth) - np.array(invested_amount), bottom=invested_amount, label='Interest Gained', color='gold')

# Add labels and title
plt.xlabel('Months')
plt.ylabel('Total Amount (INR Crores)')
plt.title('Recurring Deposit Growth: Invested Amount and Interest Gained '+ 
          str(monthly_inr)+' INR per Month with '+str(int(interest_rate*100))+'% Interest)')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

### Analysys of strategy if we do not commit to RD and try to send money whenever the rates are high


```python
# Initialize arrays to store results for each simulation
months = num_months
final_eur_values = np.zeros(num_simulations)
total_eur_spent_rd = np.zeros(num_simulations)
final_inr_no_rd = np.zeros(num_simulations)
final_eur_no_rd = np.zeros(num_simulations)
total_eur_spent_no_rd = np.zeros(num_simulations)
monthly_eur_rd = []
monthly_eur_no_rd = []
# Strategy parameters for converting when exchange rates are high
moving_window = 6  # Consider a moving average of the past 6 months as a baseline for conversion
# Parameters for the savings account interest rate
savings_interest_rate = 0.035  # 3.5% annual interest for savings account
monthly_savings_interest_rate = savings_interest_rate / 12  # Monthly interest rate for savings account

# Loop through each simulation
for i in range(num_simulations):
    # Simulate monthly EUR required for the RD strategy
    eur_spent_rd = 60000 / simulated_rates[:, i]  # EUR spent each month for RD strategy
    total_eur_spent_rd[i] = np.sum(eur_spent_rd)  # Total EUR spent for RD over 10 years
    monthly_eur_rd.append(eur_spent_rd)
    # Convert the final RD amount in INR back to EUR using the last month's exchange rate
    final_eur_values[i] = rd_growth[-1] / simulated_rates[-1, i]  # Final EUR value from RD

    # Simulate the strategy to convert at higher rates
    total_inr_converted = 0
    eur_held = 0  # Start with holding EUR
    eur_spent_no_rd = 0
    monthly_eur_spent = [0,0,0,0,0,0]
    for t in range(moving_window, months):  # Start after the first moving_window months
        # Calculate moving average of the exchange rate over the past 6 months
        moving_avg = np.mean(simulated_rates[t - moving_window:t, i])
        
        # If the current exchange rate is higher than the moving average, convert to INR
        if simulated_rates[t, i] > moving_avg:
            eur_spent_no_rd += eur_held
            monthly_eur_spent.append(eur_held)
            total_inr_converted += eur_held * simulated_rates[t, i]  # Convert all held EUR to INR
            eur_held = 0  # Reset EUR held after conversion
        else:
            eur_held += 60000 / simulated_rates[t, i]  # Hold onto EUR (add monthly amount in EUR)
            monthly_eur_spent.append(0)
        total_inr_converted *= (1 + monthly_savings_interest_rate)

    # Add any remaining EUR at the end of the period (convert at the last exchange rate)
    total_inr_converted += eur_held * simulated_rates[-1, i]  # Convert remaining EUR at last rate
    final_inr_no_rd[i] = total_inr_converted  # Final INR value without RD strategy
    final_eur_no_rd[i] = final_inr_no_rd[i] / simulated_rates[-1, i]
    total_eur_spent_no_rd[i] = eur_spent_no_rd 
    monthly_eur_no_rd.append(monthly_eur_spent)

# Determine profitability for RD strategy
profitable_rd_simulations = final_eur_values > total_eur_spent_rd

# Determine profitability for non-RD strategy (converting at high exchange rates)
profitable_no_rd_simulations = final_eur_no_rd > total_eur_spent_no_rd 

# Plotting histogram to show comparison of final INR values
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(final_eur_values, bins=30, color='pink', edgecolor='black', label='The worth of final amount')
plt.hist(total_eur_spent_rd, bins=30, color='lavender', edgecolor='black', label='The amount invested')
#plt.axvline(np.average(total_eur_spent_rd), color='lime', linestyle='--',label=f"average:{np.average(total_eur_spent_rd):.2f}")
#plt.axvline(np.average(total_eur_spent_no_rd), color='black', linestyle='--',label=f"average:{np.average(total_eur_spent_no_rd):.2f}")
#plt.hist(profitability_comparison, bins=30, color='lightcoral', edgecolor='black')
#plt.axvline(0, color='red', linestyle='--', label='Breakeven Point')
plt.xlabel('EUR')
plt.ylabel('Number of Simulations')
plt.title('Profitability Analysis for RD strategy')
plt.legend()


plt.subplot(1, 2, 2)
plt.hist(final_eur_no_rd, bins=30, color='pink', edgecolor='black',label='The worth of final amount')
plt.hist(total_eur_spent_no_rd, bins=30, color='lavender', edgecolor='black', label='The amount invested')
#plt.axvline(np.average(total_eur_spent_rd), color='lime', linestyle='--',label=f"average:{np.average(total_eur_spent_rd):.2f}")
#plt.axvline(np.average(total_eur_spent_no_rd), color='black', linestyle='--',label=f"average:{np.average(total_eur_spent_no_rd):.2f}")
#plt.hist(profitability_comparison, bins=30, color='lightcoral', edgecolor='black')
#plt.axvline(0, color='red', linestyle='--', label='Breakeven Point')
plt.xlabel('EUR')
plt.ylabel('Number of Simulations')
plt.title('Profitability Analysis for Non-RD strategy')
plt.legend()

plt.tight_layout()
plt.show()
# Calculate percentage of profitable simulations for each strategy
rd_profitability_percentage = np.sum(profitable_rd_simulations) / num_simulations * 100
non_rd_profitability_percentage = np.sum(profitable_no_rd_simulations) / num_simulations * 100

# Output the profitability percentages
print(f"Was it worth implementing this strategy?\n",
      f"Percentage of profitable simulations for RD strategy: {rd_profitability_percentage:.2f}%")
print(f" Percentage of profitable simulations for non-RD strategy: {non_rd_profitability_percentage:.2f}%")
print(f"This shows that irrespective of exchange rate, sending money is profitable")
       
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
rd_rate = 0.07 / 12  # 7% annual interest, compounded monthly for RD
savings_rate = 0.035 / 12  # 3.5% annual interest, compounded monthly for savings

# Simulation settings
num_simulations = simulated_rates.shape[1]  # Use the already simulated exchange rate paths
num_months = simulated_rates.shape[0]
rd_investment_per_month = 60000  # INR per month for RD

# Arrays to store results
rd_total_values = np.zeros(num_simulations)  # For total value at the end of RD strategy
non_rd_total_values = np.zeros(num_simulations)  # For total value at the end of Non-RD strategy

# Simulate RD strategy and Non-RD strategy
for i in range(num_simulations):
    # RD Strategy
    rd_balance = 0
    for t in range(num_months):
        rd_balance = rd_balance * (1 + rd_rate) + rd_investment_per_month  # RD grows each month

    rd_total_values[i] = rd_balance  # Final RD balance for this simulation
    
    # Non-RD Strategy
    non_rd_balance = 0
    eur_balance = 0
    for t in range(num_months):
        # Assume you convert if the exchange rate is above a certain threshold
        if simulated_rates[t, i] > 95:  # You can adjust this threshold
            non_rd_balance += eur_balance * simulated_rates[t, i]  # Convert EUR to INR
            eur_balance = 0  # Reset EUR balance
        else:
            eur_balance += 614  # Save 1000 EUR if rates are not favorable
        
        # Apply interest on INR balance
        non_rd_balance *= (1 + savings_rate)
    
    # Convert any remaining EUR balance to INR at the last month's exchange rate
    non_rd_balance += eur_balance * simulated_rates[-1, i]
    non_rd_total_values[i] = non_rd_balance  # Final Non-RD balance for this simulation

# Calculate the percentage difference between the two strategies for each simulation
percentage_differences = (non_rd_total_values - rd_total_values) / rd_total_values * 100

# Check how many simulations have a difference of <= 5%
within_threshold = np.sum(np.abs(percentage_differences) <= 5)

# Print results
print(f'Percentage of simulations where the difference between RD and Non-RD is within 5%: '
      f'{within_threshold / num_simulations * 100:.2f}%')

# Plot a comparison of total values for RD and Non-RD strategies
plt.figure(figsize=(10, 6))

plt.hist(percentage_differences, bins=50, color='skyblue', edgecolor='black')
plt.axvline(x=5, color='green', linestyle='--', label='5% Threshold')
plt.axvline(x=-5, color='green', linestyle='--')

plt.title('Distribution of Percentage Difference Between RD and Non-RD Strategies')
plt.xlabel('Percentage Difference (%)')
plt.ylabel('Frequency')
plt.legend()

plt.grid(True)
plt.show()
```
